# Limpieza
- **Nombre de las columnas** -> pasar a minúsculas y añadir (‘\_’) entre palabras.
- **Crear columnas** -> crear antigüedad del cliente en años y total de hijos

```python
#Tratamiento de datos
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Exploracion de archivos
import sys
sys.path.append('../')

#Funciones personalizadas
from src import soporte_limpieza as spl 
```

```python
# Ruta del archivo
path_customer = '../data/raw/customer-details.xlsx'

# Cargar todas las hojas en una lista de DataFrames
xls = pd.ExcelFile(path_customer)
dfs = [pd.read_excel(xls, sheet_name=sheet, index_col=0) for sheet in xls.sheet_names]

# Concatenar todas las hojas en un único DataFrame
df_o = pd.concat(dfs, ignore_index=True)

# Crear copia para trabajar y no modificar el original
df = df_o.copy()
```

## Nombre Columnas
Pasamos nombres a minúsculas y añadimos (‘\_’) entre palabras.

```python
df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))
```

```python
df = df.rename(columns={
    'kidhome': 'kids_home',
    'teenhome': 'teens_home',
    'numwebvisitsmonth': 'num_web_visits_month'
})
```

## Columnna `customer_tenure_years`
Creamos una nueva columnna para ver la antigüedad en años de los clientes.

```python
pd.Timestamp.today()
```

```python
today = pd.Timestamp.today()
df['customer_tenure_years'] = ((today - df['dt_customer']).dt.days / 365).round(1)
```

## Columna `total_children`
Creamos una nueva columna para ver el total de hijos de los clientes.

```python
df['total_children'] = df['kids_home'] + df['teens_home']
```

```python
df.to_csv('../data/processed/01.data_limpios_customer.csv', index=False)
```
