# EDA Customer dataset
En este notebook se realiza un Análisis Exploratorio de los Datos (EDA) del dataset `customer-details`, una vez finalizada la limpieza de datos.

```python
#Tratamiento de datos
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Visualizaciones
import matplotlib.pyplot as plt
import seaborn as sns

#Exploracion de archivos
import sys
sys.path.append('../')

#Funciones personalizadas
from src import soporte_visualizacion as spv 
```

```python
df_o = pd.read_csv('../data/processed/01.data_limpios_customer.csv', parse_dates=['dt_customer'])
df = df_o.copy()
```

## Distribución de la antigüedad del cliente

Analizamos como se distribuye la antigüedad de los clientes

- Variables analizadas:
    - customer_tenure_years: años desde el alta del cliente

```python
spv.plot_hist(df, cols=['customer_tenure_years'], bins=30)
```

La distribución de la antigüedad del cliente muestra una mayor concentración de clientes con muchos años en la entidad.

## Ingresos medios según número total de hijos

Analizamos el nnúmero de hijos en el hogar y el ingreso medio.

- Variables analizadas:
    - total_children
    - income

```python
plt.figure(figsize=(7, 4))
sns.boxplot(
    x='total_children',
    y='income',
    data=df,
    color='#4C72B0'
)

plt.title('Distribución de ingresos según número total de hijos', fontweight='bold')
plt.xlabel('Número total de hijos')
plt.ylabel('Ingreso')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

El ingreso medio se mantiene relativamente estable independientemente del número total de hijos.
No se observan diferencias relevantes entre grupos.

## Relación entre income y num_web_visits_month

Analizar si hay una relación lineal entre el nivel de ingresos y la frecuencia de visitas web.

- Variables analizadas:
    - income
    - num_web_visits_month


```python
spv.scatter_eda(
    df=df,
    x='num_web_visits_month',
    y='income',
    title='Relación entre visitas web mensuales e ingresos',
    xlabel='Visitas web mensuales',
    ylabel='Ingreso'
)
```

No se observa una relación lineal clara entre el nivel de ingresos y la frecuencia de visitas web

## Antigüedad del cliente vs visitas web

Analizamos si los clientes más antiguos muestran mayor visitas web

- Variables analizadas:

    - customer_tenure_years

    - num_web_visits_month

```python
spv.scatter_eda(
    df=df,
    x='customer_tenure_years',
    y='num_web_visits_month',
    title='Antigüedad del cliente vs visitas web mensuales',
    xlabel='Antigüedad del cliente (años)',
    ylabel='Visitas web mensuales'
)
```

No se observa una relación clara entre la antigüedad del cliente y el número de visitas web mensuales. La dispersión de los puntos es similar a lo largo de todos los niveles de antigüedad.
